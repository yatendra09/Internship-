```python
# Click on Trending

trending = driver.find_element_by_xpath('//*[@href="/trending"]')
try:
    driver.get(trending.get_attribute('href'))
except ElementNotInteractableException:
    driver.get(trending.get_attribute('href'))
```

```python
#scrap all names tag

names=[]
try:
    name=driver.find_elements_by_xpath("//*[@id='js-pjax-container']/div[3]/div/div[2]/article/h1/a")
    for i in name:
        names.append(i.text)
except NoSuchElementException:
    names.append('N/A')
```

```python
# scrap all description tag

desc=[]
try:
    description=driver.find_elements_by_xpath("//*[@id='js-pjax-container']/div[3]/div/div[2]/article/p")
    for i in description:
        desc.append(i.text)
except NoSuchElementException:
    desc.append('N/A')
```

```python
urls=[]
counts=[]
match_page=driver.find_elements_by_xpath('//*[@id="js-pjax-container"]/div[3]/div/div[2]/article/h1/a')
for i in match_page:
    urls.append(i.get_attribute('href'))
```

```python
#Scrap all language tag

lang=[]
try:
    for i in urls:
        driver.get(i)
        count =driver.find_element_by_xpath('//*[@id="repo-content-pjax-container"]/div/div[2]/div[2]')    
        listy=count.text.split('\n')
        if 'Languages' in listy:
            lang.append(listy[listy.index('Languages')+1])
        else:
            lang.append('N/A')
except NoSuchElementException:
    lang.append('N/A')
```

```python
counts=[]
try:
    for i in urls:
        driver.get(i)
        count =driver.find_element_by_xpath('//*[@id="repo-content-pjax-container"]/div/div[2]/div[2]')    
        listy=count.text.split('\n')
        for j in listy:
            if j.split()[0]=='Contributors':
                counts.append(j.split()[1])
                break
        else:
            counts.append('N/A')
except NoSuchElementException:
    counts.append('N/A')
```

```python
len(names),len(desc),len(lang),len(counts)
```

```python
df=pd.DataFrame()
df['Repository title']=names[0:24]
df['Repository description']=desc[0:24]
df['Language used']=lang[0:24]
df['Contributors count']=counts[0:24]
df
```
