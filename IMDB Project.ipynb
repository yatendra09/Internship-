```python
from bs4 import BeautifulSoup

import requests
```

```python
page = requests.get('https://www.imdb.com/chart/top/')
```

```python
#getting html text from webpage

soup = BeautifulSoup(page.text,'html.parser')
soup
```

```python
# first we will use html tag where we have the first title of the rasturant
first_title = soup.find('td',class_='titleColumn')
print(first_title)
```

```python
first_title.text.replace('\n','')
#replacing extra material
first_title.text.replace('\n','')
```

```python
# scraping first rating
rate = soup.find('td',class_='ratingColumn')
rate.text
rate.text.replace('\n','')
```

```python
#scraping first Realese year

yearOR = soup.find('span',class_='secondaryInfo')
yearOR.text
```

```python
#now we have all the tags in which there are the movie titles 
# now we will extract the text from these tags one bye one by looping over these tags 

movie_names =[] # empty list for store the titles

for i in soup.find_all('td',class_='titleColumn'):
    movie_names.append(i.text.replace('\n',''))
    
movie_names


```

```python
# scraping all rating

Rating= [] # empty list

for i in soup.find_all('td',class_='ratingColumn imdbRating'):
    Rating.append(i.text.replace('\n',''))
    
Rating
```

```python
#scraping year of release

Year = [] # empty list

for i in soup.find_all('span',class_='secondaryInfo'):
    Year.append(i.text.split('|')[0])
        
    
Year
```

```python
#printing legnth
print(len(movie_names),len(Year),len(Rating))
```
