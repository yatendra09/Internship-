Write a python program to scrape house details from mentioned URL. It should include house title, location,
area, EMI and price from nobroker.in.

```python
#importing required libs
from bs4 import BeautifulSoup
import requests
```

```python
#sending get request
page = requests.get('https://www.nobroker.in/property/rent/pune/Baner?searchParam=W3sibGF0IjoxOC41NjQyNDUyLCJsb24iOjczLjc3Njg1MTEsInBsYWNlSWQiOiJDaElKeTlOZDhNLS13anNSZmF0Xy01Y1NrYUUiLCJwbGFjZU5hbWUiOiJCYW5lciJ9XQ==&radius=2.0&sharedAccomodation=0&city=pune&locality=Baner')
```

```python
#checking status
page
```

```python
#gettig html tet from webpage
soup = BeautifulSoup(page.text,'html.parser')
```

```python
soup
```

```python
#scraping house_title
house_title = soup.find('h2',class_="heading-6 font-semi-bold nb__25Cl7")
house_title.text
```

```python
#scraping location
location = soup.find('div',class_="nb__1EwQz")
location.text
```

```python
#scraping area
area = soup.find('div',class_="nb__FfHqA")
area.text
```

```python
#scraping house price
price = soup.find('div',class_="nb__3q08s")
price.text.split(',')[2:4]
price2=price.text.replace('No Extra MaintenanceRent (Negotiable)','')
final_price = price2.replace('1,100 sqftBuiltup\x8250,000Deposit\x82','')
final_price
```
